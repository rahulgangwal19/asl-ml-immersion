# What is mean

Mean is the same as AVERAGE. You can find the average of a set of numbers
by ADDING them up and dividing them by HOW MANY they are.

# What is Standard Deviation
Measure of HOW SPREAD OUT numbers are

# What is Variance
Variance is the AVERAGE SQUARED DIFFERENCE OF MEAN

To work it out 

1.Work out the MEAN

2.For each number,subtract the mean and square the result.

3.Find the AVERAGE of SQUARED DIFFERENCES


```python
# Demo of variance
high_var_array = np.array([1,100,200,300,4000,5000])
low_var_array = np.array([2,4,6,8,10])

np.var(high_var_array),np.var(low_var_array)
```

```python
np.std(high_var_array),np.std(low_var_array)
```

```python
# standard deviation is the square root of variance
np.sqrt(np.var(high_var_array))
```

# Sorting Arrays

np.sort() __ sort values in a specified dimension of an array

np.argsort() __ return the indices to sort the array on a given axes

np.argmax() __ return the index/indices which gives the highest value(s)
along an axis

np.argmin() __ return the index/indices which gives the lowest value
along an axis

```python
random_array
```

```python
# Row wise
np.sort(random_array)
```

```python
np.argsort(random_array)
```

```python
a1
```

```python
# Return the indices that would sort an array
np.argsort(a1)
```

```python
# No axis
np.argmin(a1)
```

```python
random_array
```

```python
# Down the vertical
np.argmax(random_array,axis=1)
```

```python
# across the horizontal
np.argmin(random_array,axis=0)
```
